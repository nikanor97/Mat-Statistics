```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.stats as sts
import math
%matplotlib inline
```

```python
def graph(teta, N, scale_from, scale_to):
    print '\n','teta = ',teta,', N = ',10000
    uniform_rv = sts.uniform(0, teta)
    x = uniform_rv.rvs(N)
    y = range(1, N + 1)
    dif2 = list()
    dif3 = list()
    dif4 = list()
    dif5 = list()
    dif1 = abs(2 * np.cumsum(x) / range(1, N + 1) - teta)
    for i in range (1, N + 1): 
        dif2.append(abs(float(x.cumsum()[i - 1] / i + x[:i].max() / 2) - teta))
        dif3.append(abs((i + 1) * x[:i].min() - teta))
        dif4.append(abs(x[:i].min() + x[:i].max() - teta))
        dif5.append(abs(x[:i].max() * float((i + 1) / i) - teta))
    plt.figure(figsize= (15, 5))
    plt.subplot(121)
    plt.axis([0, N, scale_from, scale_to])
    plt.plot(y, dif1, label = u'первая оценка') #blue
    plt.plot(y, dif2, label = u'вторая оценка') #orange
    plt.plot(y, dif4, label = u'третья оценка') #green
    plt.plot(y, dif5, label = u'пятая оценка') #red
    plt.legend()
    plt.subplot(122)
    plt.plot(y, dif3, label = u'четвертая оценка') #blue
    plt.legend()
    plt.show()
```

```python
print(
u'В левой колонке представлены графики для первой оценки (синий цвет), для второй (оранжевый), для четвертой (зеленый), для пятой (красный).'
    u' В правой колонке представлен график для тетьей оценки.''\n')
graph(1, 10000, -0.01, 0.15)
graph(3, 10000, -0.05, 0.4)
graph(5, 10000, -0.09, 0.6)
graph(7, 10000, -0.1, 0.8)
graph(9, 10000, -0.1, 1)
print(u'Из графиков видно, что лучшими оценками при n = N являются оценки на красном и зеленом графиках, то есть четвертая и пятая.')
print(u'Вывод: четвертая и пятая оценки - лучшие')
```
