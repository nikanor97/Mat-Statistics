```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.stats as sts
import math
import statsmodels
from statsmodels.distributions.empirical_distribution import ECDF
%matplotlib inline
```

```python
teta1 = 2
uniform_rv1 = sts.uniform(0, teta1)
X = np.empty((100, 1000))
dif1 = np.empty((100, 1000)) #массив квадратичных функций потерь для первой оценки
dif2 = np.empty((100, 1000)) #массив квадратичных функций потерь для второй оценки
dif3 = np.empty((100, 1000)) #массив квадратичных функций потерь для третьей оценки
dif4 = np.empty((100, 1000)) #массив квадратичных функций потерь для четвертой оценки
for a in range(1, 101):
    X[a - 1] = uniform_rv1.rvs(1000)
for a in range(1, 101):
    for i in range(1, 1001): #заполняем массивы функций для оценок
        np.append(dif1[a - 1], ((2. * X[a - 1][i - 1] / i) - teta1)**2)
        np.append(dif2[a - 1], (((i + 1.) * X[a - 1][:i].min()) - teta1)**2)
        np.append(dif3[a - 1], ((X[a - 1][:i].min() + X[a - 1][:i].max()) - teta1)**2)
        np.append(dif4[a - 1], ((X[a - 1][:i].max() * float((i + 1) / i)) - teta1)**2)
res1 = np.empty((1000)) #массив усредненных функций по выборкам
res2 = np.empty((1000)) #массив усредненных функций по выборкам
res3 = np.empty((1000)) #массив усредненных функций по выборкам
res4 = np.empty((1000)) #массив усредненных функций по выборкам
for i in range(1, 1001):
    for a in range(1, 101):
        res1[i - 1] += dif1[a - 1][i - 1]
        res2[i - 1] += dif2[a - 1][i - 1]
        res3[i - 1] += dif3[a - 1][i - 1]
        res4[i - 1] += dif4[a - 1][i - 1]
res1 /= 100
res2 /= 100
res3 /= 100
res4 /= 100
```

```python
plt.plot(range(1, 1001), res1);
plt.title('res1 for teta')
plt.show()
plt.plot(range(1, 1001), res2);
plt.title('res2 for teta')
plt.show()
plt.plot(range(1, 1001), res3);
plt.title('res3 for teta')
plt.show()
plt.plot(range(1, 1001), res4);
plt.title('res4 for teta')
plt.show()
```
