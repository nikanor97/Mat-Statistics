```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.stats as sts
import math
%matplotlib inline
```

```python
teta = 1
N = 200000
```

```python
def func(k):
    print 'k = ',k
    expon_rv = sts.expon(scale=teta) #решаем аналогично предыдущей задаче
    x = expon_rv.rvs(N)
    xk = x ** k
    xkcumsum = xk.cumsum()
    y = range(1, N + 1)
    fack = math.factorial(k)
    dif = np.empty((N))
    for i in range (1, N + 1):
        dif[i - 1] = (abs( (1. * fack / (xkcumsum[i - 1] / i)) ** (1. / k) - teta))
    plt.plot(y, dif) #blue
    plt.axis([0, N, -0.3, 0.3])
    plt.show()
```

```python
print (u'Ниже представлены 10 графиков для 10 различных значений k.\n')
func(1)
func(2)
func(3)
func(4)
func(8)
func(10)
func(13)
func(15)
func(18)
func(20)
print(u'Вывод: наилучшим образом ведет себя оценка с параметром k = 1.')
```
